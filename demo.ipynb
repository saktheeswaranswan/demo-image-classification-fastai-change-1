```python
from fastai import *
from fastai.vision import *
from fastai.core import *
```

```python
path = Path('imgs')
```

```python
 #thankful to mr.Shuyi Wang for the code and its after some change taken from this repo
    
    #https://github.com/wshuyi/demo-image-classification-fastai
```

```python
data = ImageDataBunch.from_folder(path, test='test', ds_tfms=get_transforms(), size=224)
```

```python
data.show_batch(rows=3, figsize=(10,10))
```

```python
learn = cnn_learner(data, models.resnet34, metrics=error_rate)
learn.fit_one_cycle(1)
```

```python
preds,y,losses = learn.get_preds(with_loss=True)
interp = ClassificationInterpretation(learn, preds, y, losses)
interp.plot_confusion_matrix()
```

```python
interp.plot_top_losses(9, figsize=(10,10))
```

```python
interp.plot_confusion_matrix()
```
